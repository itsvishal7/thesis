```python
import pandas as pd
import itertools
```

```python
cp = pd.read_csv('./csvs/cp/cp.csv')
cp.tail()
```

```python
np = pd.read_csv('./csvs/np/np.csv')
np.tail()
```

```python
import os
arr = os.listdir('./csvs/pp')
print(arr)
```

```python
pp = pd.DataFrame()
for each in arr:
    df2 = pd.read_csv('./csvs/pp/'+each)
    pp = pp.append(df2)
```

```python
com = pd.merge(cp, pp, on='district', suffixes = ['_cp', '_pp'])
```

```python
col = np.columns.tolist()
col[1:] = list(map(lambda x: x + '_np', col[1:]))
np.columns = col
```

```python
np.head()
```

```python
com = pd.merge(com, np, on='district')
com.head()
```

```python
data = com.copy()
```

```python
indices = set()
groups = set()
others = set()
for c in data.columns:
    if '_' in c:
        (i, g) = c.split('_')
        c2 = pd.MultiIndex.from_tuples((i, g),)
        indices.add(int(i))
        groups.add(g)
    else:
        others.add(c)
```

```python
columns = list(itertools.product(groups, indices))
```

```python
columns = pd.MultiIndex.from_tuples(columns)
```

```python
ret = pd.DataFrame(columns=columns)
```

```python
for c in columns:
    ret[c] = data['%d_%s' % (int(c[1]), c[0])]
```

```python
for c in others:
    ret[c] = data['%s' % c]
```

```python
ret
```

```python
ret.columns = ret.columns.swaplevel(0, 1)
```

```python
ret.sort_index(axis=1, level=0, inplace=True)
```

```python
ret
```

```python
ret.to_csv('./csvs/rmse.csv', index = False)
```
